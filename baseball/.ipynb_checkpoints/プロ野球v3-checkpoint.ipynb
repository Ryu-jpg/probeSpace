````python
import re
#カテゴリカル変数の指定

# pitcherHandの変換
pitcherHand = {'L':0,'R':1}
train['pitcherHand'] = train['pitcherHand'].map(pitcherHand)
test['pitcherHand'] = test['pitcherHand'].map(pitcherHand)
# batterHandの変換
batterHand = {'L':0,'R':1}
train['batterHand'] = train['batterHand'].map(batterHand)
test['batterHand'] = test['batterHand'].map(batterHand)

#b1
b1 = {True:1,False:0}
train['b1'] = train['b1'].map(b1)
test['b1'] = test['b1'].map(b1)

#b2
b2 = {True:1,False:0}
train['b2'] = train['b2'].map(b2)
test['b2'] = test['b2'].map(b2)
#b3
b3 = {True:1,False:0}
train['b3'] = train['b3'].map(b3)
test['b3'] = test['b3'].map(b3)

# inning を数値に変換
train['inning_num'] =  train['inning'].apply(lambda x: re.sub("\\D", "", x))
test['inning_num'] =  test['inning'].apply(lambda x: re.sub("\\D", "", x))

# 表裏を判定する関数
def omote_ura(x):
  if '表' in x:
    return 0
  else:
    return 1

# 表裏の列を追加
train['inning_ForB'] =  train['inning'].apply(lambda x: omote_ura(x))
test['inning_ForB'] =  test['inning'].apply(lambda x: omote_ura(x))

#ピッチャーとバッターの型変換
test = test.astype({'pitcher': str})
test = test.astype({'batter': str})
train = train.astype({'pitcher': str})
train = train.astype({'batter': str})

# inningの削除
train = train.drop('inning',axis=1)
test = test.drop('inning',axis=1)

# speedの削除
train = train.drop('speed',axis=1)

#	ballPositionLabelの削除
train = train.drop('ballPositionLabel',axis=1)
#	pitchTypeの削除
train = train.drop('pitchType',axis=1)

#ballXの削除
train = train.drop('ballX',axis=1)
#	ballYの削除
train = train.drop('ballY',axis=1)
#dirの削除	
train = train.drop('dir',axis=1)

#distの削除	
train = train.drop('dist',axis=1)
#battingTypeの削除	
train = train.drop('battingType',axis=1)
#isOutsの削除
train = train.drop('isOuts',axis=1)


#speedの数字化
#ballPositionLabelの変換
#ballYの変換
#dirの変換
#battingTypeの変換
#isOutsの変換

#不要列の削除、gameId batter piccher
````
